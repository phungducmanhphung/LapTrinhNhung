```python
import serial
import time
import csv

com = "COM3"
baud = 115200
x = serial.Serial(port=com, baudrate=baud, timeout = 1)
```

```python
def export_sinal_to_csv(m,n,type_data):   
    for i in range(m, n):
        with open(f'output2/{type_data}_{i}.csv', mode='w', newline='') as sensor_file:
            sensor_writer = csv.writer(sensor_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
            sensor_writer.writerow(["STT","AccelX", "AccelY", "AccelZ", "GyroX", "GyroY", "GyroZ"])    
        count = 0
        while x.isOpen() == True:
            data = str(x.readline().decode()).strip()
            print(data)
            data =data.rstrip("b\\rn\''")
            # data =data[2:]
            if data is not '' :
            #  print(data[0])
                if data[0] is not '\\':
                    count = count+1
                    print(data)
                    with open(f'output2/{type_data}_{i}.csv', mode='a', newline='') as sensor_file:
                        sensor_writer = csv.writer(sensor_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
                        sensor_writer.writerow([count, *data.split(",")])
                    if count >= 200:
                        break
        
```
