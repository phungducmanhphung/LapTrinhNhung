```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import common
```

```python
X, Y = common.DocDuLieu()
X = common.ChuyenDoiTatCaDuLieuTuHangSangCot_AllFile(X)
X_fourier = []
```

```python
X_fourier = []
for arr2 in X:
    sample = []
    # lap qua cac cot
    for arr1 in arr2:
        col = common.TinhTanSoFourierCuaMang(arr1)
        sample.append(np.array(col))
    X_fourier.append(np.array(sample))

X_fourier = np.array(X_fourier)

print(X_fourier.shape)
```

```python
# plot biến đổi fourier của các cột trong các mẫu dữ liệu
x_lim_1 = 0
x_lim_2 = 200

so_mau = 2

print(len(X_fourier))
print(X_fourier[0].shape)

for i in range(len(X_fourier)):
    if i >= so_mau:
        break
    arr2 = list(X_fourier[i])
    l = len(arr2)
    for j in range(l):
        arr1 = arr2[j]
        plt.subplot(3,2,j+1)
        plt.xlim(x_lim_1, x_lim_2)
        plt.plot(arr1)
    plt.show()
```
